```python
from Bio import SeqIO
```

```python
fasta_sequences = SeqIO.parse(open('proc_prot.fasta'),'fasta')
```

```python
fasta_sequences
```

```python
with open('output.fasta', 'w') as out_file:
    for fasta in fasta_sequences:
        name, sequence = fasta.id, str(fasta.seq)
        new_id = id_dict.get(name, name)  # If no matching taxid found, keep the original protein id
        fasta.id = new_id
        fasta.description = new_id  # Set description to be the same as id
        SeqIO.write(fasta, out_file, 'fasta')
```

```python
from ete3 import Tree
```

```python
t = Tree('arch.treefile', format = 3)
for leaf in t:
    new_id = id_dict.get(leaf.name, leaf.name)  # If no matching taxid found, keep the original saccver
    leaf.name = new_id
```

```python
t.write(outfile='new_output_arch.treefile')
```

Convert tvs to itol format

```python
# Define the input and output file names
input_file = 'tax_tab_bac.tsv'  # Replace with your actual tvs file name
output_file = 'itol_annotation.txt'

# Define the header for the iTOL annotation file
itol_header = """DATASET_SIMPLE
SEPARATOR TAB
DATASET_LABEL Taxonomy
COLOR #ff0000
FIELD_COLORS #ff0000
FIELD_SHAPES 1
LEGEND_TITLE Taxonomy
LEGEND_SHAPES 1
LEGEND_COLORS #ff0000
LEGEND_LABELS Taxonomy
MARGIN 0
LABELS
"""

# Read the tvs file and write the iTOL annotation file
with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
    outfile.write(itol_header)
    for line in infile:
        outfile.write(line)

print(f"Annotation file '{output_file}' created successfully.")
```
