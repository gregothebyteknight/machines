# First

```python
# download a list of taxid in csv format with one column
taxid_list = pd.read_csv('../data/taxids.csv', sep = '\t', header = None)
taxid_list = taxid_list.rename(columns={0: "taxid"}) # naming the column with taxid
```

```python
if taxid_list.iloc[-1,-1] == "Search has CONVERGED!":
    taxid_list = taxid_list.iloc[:-1,]
```

В некоторых строчках присутствуют несколько идентификаторов taxid - из них были оставлены только первые представители ряда

```python
taxid_list["taxid"] = pd.DataFrame(taxid_list["taxid"].str.split(";").str[0].astype(int))
```

```python
def get_lineage(tax):
  """
  This function returns the lineage of a taxon given its taxid

  @tax(int): taxid of the taxon
  """
  lineage = ncbi.get_lineage(tax)
  Se_lineage = pd.Series(ncbi.get_taxid_translator(lineage), name=taxon_id)
  return ",".join(Se_lineage[lineage].values)

# add taxonomy with get_lineage function
taxid_list['taxonomy'] = taxid_list.taxid.apply(get_lineage)
```

```python
# убираем лишние данные, которые мешают правильно разбить список таксономии на столбики. если есть и другие лишние данные, их можно сюда дописать
id_taxid = taxid_list

id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('FCB group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('TACK group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Bacteroidota/Chlorobiota group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Cyanobacteriota/Melainabacteria group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Terrabacteria group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Stenosarchaea group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Bacillus cereus group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Leptospirillum sp. Group III,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Aneurinibacillus group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Rhizobium/Agrobacterium group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Rhizobium/Agrobacterium group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Rhizobium/Agrobacterium group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Rhizobium/Agrobacterium group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Caldisericota/Cryosericota group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('PVC group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('PVC group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('PVC group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('PVC group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('PVC group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('PVC group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('environmental samples,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Chlorobium/Pelodictyon group,', '')
id_taxid['taxonomy']=id_taxid['taxonomy'].str.replace('Leptolyngbya group,', '')
```

```python
# делаем таблицу ID:таксономия
tax_tab = id_taxid

tax_tab = tax_tab.join(tax_tab.taxonomy.str.split(",", expand=True, n=11),  how='left', lsuffix='_left', rsuffix='_right')
tax_tab.columns =['taxid', 'taxonomy', 0, 1, "kingdom", "phylum", "class", "order", "family", "genus", "species", 9, 10, 11]
```

# Step 2

Preparation of Clustal file with taxid as labels

```python
# загружаем список taxid в видe csv с одним столбиком
taxid_list = pd.read_csv('../data/taxids.csv', sep = '\t', header = None)
```

```python
taxid_list = taxid_list.rename(columns={0: "saccver", 1: "taxid", 2: "kingdom"}) # naming the column with taxid
if taxid_list.iloc[-1,-1] == "Search has CONVERGED!":
    taxid_list = taxid_list.iloc[:-1,]
```

```python
taxid_list
```

```python
taxid_list["kingdom"].value_counts()
```

```python
# filter by kingdom
taxid_list = taxid_list[taxid_list["kingdom"] == "Bacteria"]
```

```python
# selection only saccver and taxid columns
taxid_list = pd.DataFrame(taxid_list[["saccver", "taxid"]])
```

```python
# cleaning taxid column
taxid_list["taxid"] = pd.DataFrame(taxid_list["taxid"].str.split(";").str[0])
```

```python
taxid_list
```

```python
id_dict = taxid_list.set_index('saccver')['taxid'].to_dict()
```

```python
id_dict
```

```python
fasta_sequences = SeqIO.parse(open('proc_prot.fasta'),'fasta')
```

```python
fasta_sequences
```

```python
with open('output.fasta', 'w') as out_file:
    for fasta in fasta_sequences:
        name, sequence = fasta.id, str(fasta.seq)
        new_id = id_dict.get(name, name)  # If no matching taxid found, keep the original protein id
        fasta.id = new_id
        fasta.description = new_id  # Set description to be the same as id
        SeqIO.write(fasta, out_file, 'fasta')
```

```python
t = ete3.Tree('arch.treefile', format = 3)
for leaf in t:
    new_id = id_dict.get(leaf.name, leaf.name)  # If no matching taxid found, keep the original saccver
    leaf.name = new_id
```

```python
t.write(outfile='new_output_arch.treefile')
```
